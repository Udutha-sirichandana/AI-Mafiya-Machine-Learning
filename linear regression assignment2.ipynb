```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
```

```python
test_csv=pd.read_csv(r"C:\Users\DELL\Downloads\Test.csv")
train_csv=pd.read_csv(r"C:\Users\DELL\Downloads\Train.csv")
print(train_csv,train_csv.shape,type(train_csv),test_csv,test_csv.shape,type(test_csv))
```

```python
x1=train_csv['feature_1'].values.reshape((-1,1))
x2=train_csv['feature_2'].values.reshape((-1,1))
x3=train_csv['feature_3'].values.reshape((-1,1))
x4=train_csv['feature_4'].values.reshape((-1,1))
x5=train_csv['feature_5'].values.reshape((-1,1))
x=train_csv[['feature_1','feature_2','feature_3','feature_4','feature_5']].values
y=train_csv['target'].values.reshape((-1,1))
```

```python
plt.figure(figsize=(10,5))
plt.scatter(x1,y,color="red",label='feature_1')
plt.scatter(x2,y,color="green",label='feature_2')
plt.scatter(x3,y,color="black",label='feature_3')
plt.scatter(x4,y,color="blue",label='feature_4')
plt.scatter(x5,y,color="violet",label='feature_5')
plt.xlabel('X')
plt.ylabel('Y')
plt.legend()
plt.show()

```

```python
reg=LinearRegression()
reg.fit(x,y)
```

```python
output=reg.predict(x)
output
```

```python
bias=reg.intercept_
coef=reg.coef_
print(bias,coef)
```

```python
reg.score(x,y)
```

```python
plt.figure(figsize=(10,5))
plt.plot(x,output,label='predictons')
plt.xlabel('X')
plt.ylabel('Y')
plt.legend()
plt.show()

```

```python
t1=test_csv['feature_1'].values.reshape((-1,1))
t2=test_csv['feature_2'].values.reshape((-1,1))
t3=test_csv['feature_3'].values.reshape((-1,1))
t4=test_csv['feature_4'].values.reshape((-1,1))
t5=test_csv['feature_5'].values.reshape((-1,1))
t=test_csv[['feature_1','feature_2','feature_3','feature_4','feature_5']].values
print(t,t.shape,type(t),sep='\n')
```
