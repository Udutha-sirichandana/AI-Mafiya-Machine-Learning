```python
def distance(x1,x2):
    return np.sqrt(sum((x1-x2)**2))
```

```python
def knn(X,Y,q_p,k=7):
    vals=[]
    m=X.shape[0]
    for i in range(m):
        dis=distance(q_p,X[i])
        vals.append((dis,Y[i]))
    vals = sorted(vals)
    # Nearest/First K points
    vals = vals[:k]
    vals = np.array(vals)
    new_vals=np.unique(vals[:,1],return_counts=True)
    print(new_vals)
    index = new_vals[1].argmax()
    pred = new_vals[0][index]
    
    return pred
```

```python
x = knn(X,Y,[2,2])
```
